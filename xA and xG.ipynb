```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://github.com/vaastav/Fantasy-Premier-League/blob/master/data/2021-22/understat/understat_player.csv?raw=true'
df = pd.read_csv(url)
print(df.head())
```

```python
print(df.dtypes)
```

```python
df.sort_values(by=['xG'], ascending=False)
```

```python
df.sort_values(by=['goals'], ascending=False)
```

```python
df.sort_values(by=['xA'], ascending=False)
```

```python
df.sort_values(by=['assists'], ascending=False)
```

```python
df_5goals = df[df.goals >9]
print(df_5goals)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="goals", y="xG", data=df_5goals)
plt.title("xG vs goals") #title
plt.xlabel("goals") #x label
plt.ylabel("xG") #y label
# plt.figure(figsize=(20,20))


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(df_5goals["goals"], df_5goals["xG"], df_5goals["player_name"], plt.gca()) 

plt.show()
```

```python
df_5assists = df[df.assists >6]
print(df_5assists)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="assists", y="xA", data=df_5assists)
plt.title("xA vs assists") #title
plt.xlabel("assists") #x label
plt.ylabel("xA") #y label
# plt.figure(figsize=(20,20))


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(df_5assists["assists"], df_5assists["xA"], df_5assists["player_name"], plt.gca()) 

plt.show()
```
