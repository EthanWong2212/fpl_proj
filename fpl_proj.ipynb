```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set matplotlib inline
%matplotlib inline
```

```python
url = "https://fantasy.premierleague.com/api/bootstrap-static/"
```

```python
r = requests.get(url)
json = r.json()
```

```python
json.keys()
```

```python
elements_df = pd.DataFrame(json['elements'])
```

```python
elements_df.head()
```

```python
elements_df.columns
```

```python
elements_df = elements_df[['second_name', 'element_type', 'now_cost', 'transfers_in', 'form', 'selected_by_percent', 'total_points', 'value_form', 'value_season', 'clean_sheets' ]]
elements_df.head()
```

```python
# convert strings to float
elements_df[['form', 'selected_by_percent', 'value_form', 'value_season']] = elements_df[['form', 'selected_by_percent', 'value_form', 'value_season']].astype(float)
```

```python
group_by_e_df = elements_df[elements_df['total_points'] > 0]
group_by_df = group_by_e_df.groupby('element_type').agg({'value_season': np.mean, 'total_points': np.sum})
group_by_df.head()
```

```python
elements_df.sort_values('value_season', ascending = False, inplace = True)
elements_df.head(11)
```

```python
elements_df.sort_values('total_points', ascending = False, inplace = True)
elements_df.head(20)
```

```python
# filter_elements_df.sort_values('selected_by_percent', ascending = False, inplace = True)
# separate by positions
st_df = elements_df[elements_df['element_type'] == 4]
mid_df = elements_df[elements_df['element_type'] == 3]
def_df = elements_df[elements_df['element_type'] == 2]
gk_df = elements_df[elements_df['element_type'] == 1]
```

```python
def_df.sort_values('total_points', ascending = False, inplace = True)
def_df.head(30)
```

```python
mid_df.sort_values('total_points', ascending = False, inplace = True)
mid_df.head(40)
```

```python
def_df.sort_values('total_points', ascending = False, inplace = True)
def_df.head(20)
```

```python
gk_df.sort_values('total_points', ascending = False, inplace = True)
gk_df.head(10)
```

```python
def process_total_by_val(points, cost):
    return points/cost
```

```python
st_df['total_by_val'] = np.vectorize(process_total_by_val)(st_df['total_points'], st_df['now_cost'])
st_df.sort_values('value_season', ascending = False)
st_df.head(10)
```

```python
teams_df = pd.DataFrame(json['teams'])
```

```python
teams_df.head()
```

```python
teams_df.columns
```

```python
teams_df = teams_df[['short_name', 'form', 'strength_defence_home', 'strength_defence_away', 'strength']]
teams_df.head()
```

```python
teams_df.sort_values('strength_defence_home')
```

```python
def avr_defense(home, away):
    return (home + away) / 2
```

```python
teams_df['avr_def'] = np.vectorize(avr_defense)(teams_df['strength_defence_home'], teams_df['strength_defence_away'])
teams_df.sort_values('avr_def', ascending = False, inplace = True)
teams_df.head(10)
```
